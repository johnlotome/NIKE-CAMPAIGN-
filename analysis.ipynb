```
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```
sort_leader_score = leaders.sort_values(by='Popularity_Score', ascending=False)
top10_leaders_score = sort_leader_score.iloc[:9]

plt.rcdefaults()
fig, ax = plt.subplots()


name = top10_leaders_score['screen_name']
score = top10_leaders_score['Popularity_Score']/1000


ax.barh(name, score, align='center')
ax.set_yticks(name)
ax.set_yticklabels(name)
ax.invert_yaxis()  
ax.set_xlabel('Popularity Score in Thousandths')
ax.set_title('Top Popularity Score of African Leaders')
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

plt.show()
```

```
sort_leaders_reach = leaders.sort_values(by='reach_score', ascending=False)
top10_leaders_reach = sort_leaders_reach.iloc[:9]


plt.rcdefaults()
fig, ax = plt.subplots()


name = top10_leaders_reach['screen_name']
score = top10_leaders_reach['reach_score']/1000


ax.barh(name, score, align='center')
ax.set_yticks(name)
ax.set_yticklabels(name)
ax.invert_yaxis()  
ax.set_xlabel('Reach Score in Thousandths')
ax.set_title('Top African leaders Reach Score')
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

plt.show()
```

```
sort_influencers_score = influencers.sort_values(by='Popularity_Score', ascending=False)
top10_influencers_score = sort_influencers_score.iloc[:9]


plt.rcdefaults()
fig, ax = plt.subplots()


name = top10_influencers_score['screen_name']
score = top10_influencers_score['Popularity_Score']/1000


ax.barh(name, score, align='center', color=np.random.rand(1,3))
ax.set_yticks(name)
ax.set_yticklabels(name)
ax.invert_yaxis()  
ax.set_xlabel('Popularity Score in Thousandths')
ax.set_title('Top Popularity Score of African Influencers')
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

plt.show()
```
